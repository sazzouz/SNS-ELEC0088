```python
import pandas as pd
from pandas_profiling import ProfileReport
from data import get_ctp, get_mobility
```

```python
ctp_df = get_ctp()
ctp_df
```

```python
mobility_df = get_mobility()
mobility_df
```

# Feature Selection

```python
start_date = '2020-03-20'
end_date = '2020-12-31'
```

```python
ctp_features = [
    'date',
    'state',
    'death',
    'deathIncrease',
    'hospitalized',
    'hospitalizedIncrease',
    'positive',
    'positiveIncrease',
    'negative',
    'negativeIncrease'
]
ctp_df = ctp_df[ctp_features]
ctp_df
```

```python
mobility_features = [
    'date',
    'sub_region_1',
    'sub_region_2',
    'retail_and_recreation_percent_change_from_baseline',
    'grocery_and_pharmacy_percent_change_from_baseline',
    'parks_percent_change_from_baseline',
    'transit_stations_percent_change_from_baseline',
    'workplaces_percent_change_from_baseline',
    'residential_percent_change_from_baseline'
]
mobility_df = mobility_df[mobility_features]
mobility_df
```

# Florida

```python
fl_ctp_df = ctp_df[ctp_df['state']=='FL']
fl_ctp_df = fl_ctp_df[:323]
fl_ctp_df = fl_ctp_df[::-1].reset_index(drop=True) # Reverse the order so it is in chronological order.
fl_ctp_df['date'] = pd.to_datetime(fl_ctp_df['date'])
mask = (fl_ctp_df['date'] >= start_date) & (fl_ctp_df['date'] <= end_date)
fl_ctp_df = fl_ctp_df.drop(['state'], 1)
fl_ctp_df = fl_ctp_df.loc[mask].reset_index(drop=True)
# fl_ctp_df = fl_ctp_df.set_index('date')
fl_ctp_df = fl_ctp_df.reset_index(drop=True)
fl_ctp_df
```

```python
fl_mobility_df = mobility_df[mobility_df['sub_region_1']=='Florida']
fl_mobility_df = fl_mobility_df[fl_mobility_df['sub_region_2'].isnull()] # Get for all counties (null value in sub region 2)
# fl_mobility_df = fl_mobility_df.set_index('date')
# fl_mobility_df['date'] = pd.to_datetime(fl_mobility_df['date'])
# mask = (fl_mobility_df['date'] >= start_date) & (fl_mobility_df['date'] <= end_date)
# fl_mobility_df = fl_mobility_df.loc[mask].reset_index(drop=True)
fl_mobility_df = fl_mobility_df[35:321] # Issue with Date sorting, just clipped to the correct number of rows.
fl_mobility_df = fl_mobility_df.drop(['date', 'sub_region_1','sub_region_2'], 1)
fl_mobility_df = fl_mobility_df.reset_index(drop=True) # To prevent adding rows after concating
fl_mobility_df
```

```python
fl_full_df = pd.concat([fl_ctp_df, fl_mobility_df], axis=1)
fl_full_df
```

# Profile

```python
profile = ProfileReport(fl_full_df, title='The Florida Dataset', explorative=True)
profile.to_widgets()
```
